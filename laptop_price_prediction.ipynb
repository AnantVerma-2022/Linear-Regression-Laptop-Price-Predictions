```python
# Removing column 'laptop_ID' as it doesn't affect much the price and unique for each laptop

# Machine learning models learn by finding patterns and correlations between features and the target variable. If every single row has a different 
# value (e.g., user_id_1, user_id_2, ...), the model cannot find a pattern. A column where every entry is unique is essentially a "constant" feature 
# in terms of predictive mapping—it tells the model nothing about how to generalize to new, unseen data

dataset.drop('laptop_ID', axis = 1, inplace = True)

dataset.head()
```

**Observations**

It is clear visible from the above data that features have alot text and for prediction we need numerical data. So to do so will need to do alot data(text) preprocessing.

```python
# Top 5 valuse/sample of dataframe

dataset.head()
```

- The column "Ram" and "Weight" is very easy to comvert into numeric -  so if somehow we can remove the text associated with each values of these columns then we can.

```python
# Let's start with feature 'RAM' and 'Weight' : from the observations if we remove GB and kg the easialy can be converted into int.

dataset['Ram'] = dataset['Ram'].str.replace('GB','').astype('int32')

dataset['Weight'] = dataset['Weight'].str.replace('kg','').astype('float32')
```

```python
# Check the update dataframe

dataset.head()
```

```python
# Check dtype

dataset.info()
```

```python
# In the Ram column Ram is in GB and In weight column weight in KG 
# So, Rename the 'Ram' -- 'Ram_GB' and 'Weight' -- 'Weight_Kg'
dataset.rename(columns = {'Ram':'Ram_GB','Weight':'Weight_KG'}, inplace = True)

dataset.head()
```

```python
# From above analysis, it is clear that 'Price' feature has some skewness in data. So let's analyse the Feature 'Price_euro'

plt.figure(figsize = (12,6))
sns.distplot(dataset['Price_euros'], color = 'red')
plt.xlabel('Price')
plt.title('Price Distribution')
plt.show()
```

Skewness of 0: The data is perfectly symmetrical (e.g., a normal distribution).


Positive Skewness (> 0): The distribution is "right-skewed," meaning the right tail is longer, and more weight is in the left tail of the distribution.


Negative Skewness (< 0): The distribution is "left-skewed," meaning the left tail is longer, and more weight is in the right tail of the distribution.
